### Load Seaborn and Pandas Libaries, along with the dataset called "mpg" using the `load.dataset` command

```python
# Import Library
import seaborn as sns
import pandas as pd

# Load the example mpg.csv dataset
# Reference for this command is: https://seaborn.pydata.org/generated/seaborn.load_dataset.html

mpg = pd.read_csv('mpg.csv')
mpg
```

```python
# Take a moment become familiar with the dataset using your favorite Python Pandas exploratory commands.

mpg.head()
```

```python
mpg.shape
```

```python
mpg.describe()
```

### Set the Theme

Use this reference link for set_theme selections: https://seaborn.pydata.org/generated/seaborn.set_theme.html <br>
For the example below using the style of white will match the gallery example.
<br>


```python
# Set Theme

sns.set_theme(style = "white")
```

Use `sns.set_theme(style="white")`

### Plot miles per gallon against horsepower with other characteristics as shown above. 

 Use the Seaborn chart type command called `relplot`.  Adjust the chart parameters for hue, size, sizes, alpha, palette, height, and data.<br> Details for the parameters can be found at https://seaborn.pydata.org/generated/seaborn.relplot.html

```python
# sns.relplot()

sns.relplot(data=mpg, x="horsepower", y="mpg", hue="origin",
           size="weight", sizes=(20, 200), alpha=0.7, palette="viridis", height=6)
```

```python
# Change the variables and create a different version of the same chart!

sns.relplot(
    data=mpg, x="horsepower", y="mpg", hue="acceleration", size="cylinders",
    sizes=(10, 200), alpha=0.7, palette="viridis", height=6
)
```

### For the next two visualizations, load the provided "tips" dataset. 

Seaborn provides the `load_dataset` command to accomplish this task.

```python
# Load the example tips dataset

tips = pd.read_csv('tips.csv')
tips
```

```python
# Take a moment become familiar with the dataset using your favorite Python Pandas exploratory commands.

tips.head()
```

### Set the Theme

Previously in this notebook, the theme was set to `white` let's try another one.
For the example below using the style of `whitegrid` will match the example.

```python
# Set a new Theme

sns.set_theme(style="whitegrid")
```

Use `sns.set_theme(style="whitegrid")`

### Reproduce the chart below from the Seaborn Gallery
<br>
<img src="violinplots.png">

 Use the Seaborn chart type command called `violinplot`.  Adjust the chart parameters to match the illustration above.ðŸ™„<br> Split the Violin Plot with `despin` to split the violin.  Details for the parameters can be found at https://seaborn.pydata.org/generated/seaborn.violinplot.html

```python
# Create the Violin Plot using violinplot chart type with despin to split the violin.

sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, split=True, inner="quart",
    palette="muted", height=6
)
```

```python
# Change a component of the parameters or selected dimensions and recreate.

sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, split=True,
    inner="quart", palette="Set2", height=6
)
```

### Reproduce the chart below from the Seaborn Gallery
<br>
<img src="lr_with_marg_distr.png">

Recreate the Regression Seaborn chart type command called `jointplot`.  Adjust the chart parameters to match the illustration above.ðŸ™„<br> Apply the following settings for xlim and ylim: `xlim=(0, 60), ylim=(0, 12)`.  Details for the parameters can be found at https://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
# Recreate the chart above contrasting the total_bill amount and tips given

sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg",      
    xlim=(0, 60), ylim=(0, 12), height=8
)
```

```python
# Modify one or more of the selected parameters to create a new version of this chart!

sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex",      
    xlim=(0, 60), ylim=(0, 12), height=8
)
```
